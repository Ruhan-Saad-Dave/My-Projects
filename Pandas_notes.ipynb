Long to wide table

```python
no2 = air_quality[air_quality["parameter"] == "no2"]
no2_subset = no2.sort_index().groupby(["location"]).head(2)  #Usunf 1st 2 measurements of each location while focusing on no2
no2_subset
```

```python
#To show values of 3 stations as separate columns next to each other.
no2_subset.pivot(columns = "location", values = "value")
```

Pivot table

```python
#To show mean concentration of 2 parameters in each station in table form
air_quality.pivot_tabel(values = "value", index = "location", columns = "parameter", aggfunc = "mean")

#In pivot(), the data is only rearranged.
#In pivot_table(), multiple values can be aggregated
```

```python
#When interested in row/column margis (subtotals) for each variable
air_quality.pivot_table(
    values = "value",
    index = "location",
    columns = "parameter",
    aggfunc = "mean",
    margins = True,
)

#Same results from:
air_quality.groupby(["parameter", "location"])[["values"]].mean()
```

Wide to long format

```python
#Add new index using reset_index()
no2_pivoted = no2.pivot(columns = "location", values = "value").reset_index()
no2_pivoted.head()
```

```python
#To collect all air quality no2 measurements in a single column (long format)
no_2 = no2_pivoted.melt(id_vars = "date.utc")  #Coverts wide format to long format
no_2.head()
```

```python
no_2 = no2_pivoted.melt(
    id_vars = "date.utc",
    value_vars = ["BETR801", "FR04014", "London Westminster"],
    var_name = "NO_2",
    value_name = "id_location",
)
no_2.head()

#value_vars defines which columns to melt together
#value_name provides a custom column name for the values column instead of the default column name value
#var_name provides a custom column name for the column collecting the column header names, else takes the index name or a default variable
```
